```python
x=df[['Age','EstimatedSalary']].values
x[0:5]
```

```python
y=df[['Purchased']]
y[0:5]
```

```python
#train_test_split dataset 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=4)
print("train",x_train.shape, y_train.shape)
print("test",x_test.shape,y_test.shape)
```

```python
#data processing into standard scaler

from sklearn import preprocessing
x=preprocessing.StandardScaler().fit(x).transform(x)
x[0:5]
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
print("train",x_train)
print("test",x_test)
```

```python
#modeling 
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=100,criterion='entropy',random_state=4)
classifier
```

```python
#fit the question and answer
classifier.fit(x_train,y_train)
```

```python
#predict
predi=classifier.predict(x_test)
predi
```

```python
#find the acuracy
from sklearn.metrics import f1_score
print("accuracy",f1_score(predi,y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(predi,y_test))
```
